# Matplotlib example workflow

```python
# 0. impot matplotlib ang get it ready for the ploting in Jupyter
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare date
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize=(10, 10)) #(width, hight)

# 3. Plot data
ax.plot(x, y)

# 4. Costomize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
# 5.Save & show (Yoy save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

# Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
# plot to each diffrent axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
 # Subplote option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                     figsize=(10,5))
# Plot to each diffrent indes
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

### which one shoud we use ?(pyplot vs matplotlib oo method)

* When plotting somthing qyickly, okey to use pyplot method
* WHen ploting somthoing more Advance so we will use oo mehtod

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# Pyplot method (plottinng directly)
over_50.plot(kind='scatter',
             x='age',
              y='chol',
              c='target') # c is for giving color 
plt.show()
```

```python
# OO method mix with pyplot methos
fix,ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
#ax.set_xlim([45,100]) # this will set the x limit from 45 to 100
plt.show()
```

```python
## oo method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50["chol"],
                    c=over_50["target"])
# Costomize the plot
ax.set(title="Heart Disease and Cholestrol Level",
      xlabel='Age',
      ylabel='Cholestrol')
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add Horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
plt.show()
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex= True)  #sharex = True use for if both x axis use common value of age and make looks graph cleaner then


# Adding data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50['chol'],
                     c=over_50["target"])
# Costomise ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
      # xlabel="Age",
       ylabel="Cholestrol")
# Adding a alegend to ax0
ax0.legend(*scatter.legend_elements(), title='target')

# Add a Meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# Costomising ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       #xlabel='Age',
       ylabel="Max Heart Rate")
# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="target")
# Adding meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight ="bold");

plt.show()
```

```python
# set the style
plt.style.use('seaborn-v0_8-whitegrid')
## oo method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter')  #  This changes the color scheme
# Costomize the plot
ax.set(title="Heart Disease and Cholestrol Level",
      xlabel='Age',
      ylabel='Cholestrol')
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add Horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
plt.show()
```

```python
# Costomising the y and x axis limitations
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex= True)  #sharex = True use for if both x axis use common value of age and make looks graph cleaner then


# Adding data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50['chol'],
                     c=over_50["target"],
                     cmap="winter")
# Costomise ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
      # xlabel="Age",
       ylabel="Cholestrol")
# Change the x-axis limits
ax0.set_xlim([50, 80])
# Adding a alegend to ax0
ax0.legend(*scatter.legend_elements(), title='target')

# Add a Meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
# Costomising ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       #xlabel='Age',
       ylabel="Max Heart Rate")
# Change x1 x-axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])
# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="target")
# Adding meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight ="bold");

plt.show()
```
